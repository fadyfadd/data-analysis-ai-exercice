```python
# set up
import pandas as pd
from IPython.display import display

pd.set_option('display.max_rows', 20)
shooting_agencies = pd.read_csv('fatal-police-shootings-agencies.csv')
fatal_shooting = pd.read_csv('fatal-police-shootings-data.csv')
```

```python
# shooting per state raw
shooting_per_state = fatal_shooting.groupby(['state', 'county']).agg(incident_count=('county', 'size')).reset_index()
```

```python
# shooting per state order by
shooting_per_state_order_by = shooting_per_state.sort_values(by='incident_count', ascending=False)
display(shooting_per_state_order_by)
```

```python
# county with most shootings
max_row = shooting_per_state.loc[shooting_per_state['incident_count'].idxmax()]
display(max_row)
```

```python
# nbr of shooting exceeding 150
high_shooting = shooting_per_state[shooting_per_state['incident_count'] > 150].sort_values(by='incident_count', ascending=False)
display(high_shooting)
```

```python
full_data = pd.merge(
    shooting_per_state,
    shooting_agencies,
    left_on=['state'],
    right_on=['state'],
    how='left'
).sort_values(by='incident_count', ascending=False)
selected_data = full_data[['state', 'county', 'incident_count']]
display(selected_data)
```
